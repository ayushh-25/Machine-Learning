d) Estimate the regularized least squared polynomial regression model of order k = 1,2, 3, 7, using the training set T. 
For example,

        (i) For k=1 , we need to estimate F(x) = w1x + b
        (ii) For k = 2, F(x) = w2x2 + w1x + b.

```python
def A_matrix(order,X):
    m = order+1
    n = len(X)
    A = np.zeros((n,m))
    for i in range(len(X)):
        for j in range(m):
            A[i,j] = X[i]**j
    return A

def RegularizedLeastSqPolynomialRegression(order,training_set):
    X = [i for i,j in training_set]
    Y = [j for i,j in training_set]
    A = A_matrix(order,X)
    I = np.identity(order+1)
    Lambda = 2**(-7)
    ans = np.dot(np.linalg.inv(np.dot(A.T,A)+Lambda*I),np.dot(A.T,Y))
    return ans

#M = 1
ans_m1 = RegularizedLeastSqPolynomialRegression(1,Training_set)
print('\nRegularized Least Square polynomial regression model of order M = 1: \n', ans_m1)

#M = 2
ans_m2 = RegularizedLeastSqPolynomialRegression(2,Training_set)
print('\nRegularized Least Square polynomial regression model of order M = 2: \n', ans_m2)

#M = 3
ans_m3 =  RegularizedLeastSqPolynomialRegression(3,Training_set)
print('\nRegularized Least Square polynomial regression model of order M = 3: \n', ans_m3)

#M = 7
ans_m7 = RegularizedLeastSqPolynomialRegression(7,Training_set)
print('\nRegularized Least Square polynomial regression model of order M = 7: \n', ans_m7)
```

f) Obtain the prediction on testing set and compute the RMSE for regularized least squared polynomial regression models for order k=1,2,3 and 7.

```python
def predicted(order, testing_set, ans):
    X = [i for i,j in testing_set]
    m = order+1
    predicted = []
    for k in range(len(X)):
        sum = 0
        for i in range(m):
            temp = ans[i]*(X[k]**i) 
            sum = sum + temp
        predicted.append(sum)
    return predicted

def RMSE(predicted,testing_set):
    actual = [j for i,j in testing_set]
    n = len(actual)
    RMSE = 0
    for i in range(n):
        temp = (predicted[i]-actual[i])**2
        RMSE = RMSE + temp
    RMSE = (RMSE/n)**0.5
    return RMSE
```

```python
def second_norm(*args):
    sum = 0
    for i in args:
        sum = sum + (i**2)
    return (sum**0.5)

#Training set:
xi1_200 = np.random.uniform(0,1,200)
xi2_200 = np.random.uniform(0,1,200)
ei_200 = np.random.normal(0,0.25,200)
yi_200 = np.sin(2*3.14*second_norm(xi1_200,xi2_200))+ei_200
Training_set_200 = list(zip(xi1_200,xi2_200,yi_200))

Training_set_200
```

b) Obtain the prediction on testing set and compute the RMSE for regularized least squared polynomial regression models for order k =1,2 and 3. Also plot the estimated function and target function for the training set and testing set.

```python
def combinations_matrix_of_bivariate(order,training_set):
    x1 = np.array([i for i,j,k in training_set])
    x2 = np.array([j for i,j,k in training_set])
    a = []
    for k in range(order+1):
        for i in range(order+1-k):
            for j in range(order+1-k):
                if i+j == order-k:
                    temp = (x1**i)*(x2**j)
                    a.append(temp)
    return np.array(a).T

def bi_RegularizedLeastSqPolynomialRegression(order,training_set):
    Y = [k for i,j,k in training_set]
    A = combinations_matrix_of_bivariate(order,training_set)
    I = np.identity(len(A.T))
    Lambda = 2**(-7)
    ans = np.dot(np.linalg.inv(np.dot(A.T,A)+Lambda*I),np.dot(A.T,Y))
    return ans


#M = 1
bi_ans_m1 = bi_RegularizedLeastSqPolynomialRegression(1,Training_set_200)
print('\nRegularized Least Square polynomial regression model for bivariate data of order M = 1: \n', bi_ans_m1)

#M = 2
bi_ans_m2 = bi_RegularizedLeastSqPolynomialRegression(2,Training_set_200)
print('\nRegularized Least Square polynomial regression model for bivariate data of order M = 2: \n', bi_ans_m2)

#M = 3
bi_ans_m3 =  bi_RegularizedLeastSqPolynomialRegression(3,Training_set_200)
print('\nRegularized Least Square polynomial regression model for bivariate data of order M = 3: \n', bi_ans_m3)

```

```python
def bi_predicted(order,testing_set,ans):
    A = combinations_matrix_of_bivariate(order,testing_set)
    return A@ans

def bi_RMSE(predicted,testing_set):
    actual = [k for i,j,k in testing_set]
    n = len(actual)
    RMSE = 0
    for i in range(n):
        temp = (predicted[i]-actual[i])**2
        RMSE = RMSE + temp
    RMSE = (RMSE/n)**0.5
    return RMSE

#M = 1
bi_pred_training_m1 = bi_predicted(1,Training_set_200,bi_ans_m1)
bi_rmse_training_m1 = bi_RMSE(bi_pred_training_m1,Training_set_200)
print('\nTrain RMSE for polynomial regression models for order M = 1 : ', bi_rmse_training_m1)

#M = 2
bi_pred_training_m2 = bi_predicted(2,Training_set_200,bi_ans_m2)
bi_rmse_training_m2 = bi_RMSE(bi_pred_training_m2,Training_set_200)
print('\nTrain RMSE for polynomial regression models for order M = 2 : ', bi_rmse_training_m2)

#M = 3
bi_pred_training_m3 = bi_predicted(3,Training_set_200,bi_ans_m3)
bi_rmse_training_m3 = bi_RMSE(bi_pred_training_m3,Training_set_200)
print('\nTrain RMSE for polynomial regression models for order M = 3 : ', bi_rmse_training_m3)

```

```python
import pandas as pd

motorcycle_data = pd.read_csv(r'G:\My Drive\MSc DS sem2\ML\Lab Assignments\motorcycle - motorcycle.csv')
X = np.array(motorcycle_data['x'])
Y = np.array(motorcycle_data['y'])
motorcycle_data = list(np.array(motorcycle_data))
```

```python
def A_matrix_sigmoidal(order,training_set):
    X = [i for i,j in training_set]
    m = order+1
    n = len(X)
    w = []
    b = []
    for k in range(order):
        tempW = np.random.uniform(-2,2)
        tempB = np.random.uniform(-5,5)
        w.append(tempW)
        b.append(tempB)
        
    A = np.ones((n,m))
    for i in range(len(X)):
        for j in range(m-1):
            A[i,j] = float(1/(1+(np.exp(-((w[j]*X[i])+b[j])))))
    return A

def RegularizedLeastSqPolynomialRegression_sigmoidal(order,training_set):
    X = [i for i,j in training_set]
    Y = [j for i,j in training_set]
    A = A_matrix_sigmoidal(order,training_set)
    I = np.identity(order+1)
    Lambda = 2**(-7)
    ans = np.dot(np.linalg.inv(np.dot(A.T,A)+Lambda*I),np.dot(A.T,Y))
    return ans

def predicted_sigmoidal(order,testing_set,ans):
    A = A_matrix_sigmoidal(order,testing_set)
    return np.dot(A,ans)

def RMSE_sigmoidal(predicted,testing_set):
    actual = [j for i,j in testing_set]
    n = len(actual)
    RMSE = 0
    for i in range(n):
        temp = (predicted[i]-actual[i])**2
        RMSE = RMSE + temp
    RMSE = (RMSE/n)**0.5
    return RMSE
```

II. For M =10, find the minimum mean and standard deviations of RMSE using leave-one out method by tuning the parameter λ.

```python
def leave_one_out(dataset):
    subsets = []
    for i in range(len(dataset)):
        left_out = dataset[i]
        subset = dataset[:i] + dataset[i+1:]
        subsets.append((subset, left_out))
    return subsets

def leave_one_out_test_matrix(order,test):
    X, Y = test
    m = order+1
    n = 1
    w = []
    b = []
    for k in range(order):
        tempW = np.random.uniform(-2,2)
        tempB = np.random.uniform(-5,5)
        w.append(tempW)
        b.append(tempB)
        
    A = np.ones((n,m))
    for j in range(m-1):
            A[0,j] = float(1/(1+(np.exp(-((w[j]*X)+b[j])))))   
    return A

def leave_one_out_RMSE(predicted, test):
    X, actual = test
    RMSE = ((predicted-actual)**2 )**0.5
    return RMSE

Rmse = []
dataset = leave_one_out(motorcycle_data)
for i in range(len(dataset)):
    data = dataset[i]
    train = dataset[i][0]
    test = dataset[i][1]
    coeff = RegularizedLeastSqPolynomialRegression_sigmoidal(10,train)  # ƛ = 2^(-7)
    prediction = leave_one_out_test_matrix(10,test)@coeff
    rmse = leave_one_out_RMSE(prediction,test)
    Rmse.append(rmse)

print('minimun RMSE using leave one out method: ', np.min(Rmse))
print('mean of RMSE using leave one out method: ', np.mean(Rmse))
print('SD of RMSE using leave one out method: ', np.std(Rmse))

```
