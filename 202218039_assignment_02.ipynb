• Generate 20 real number for the variable X from the uniform distribution U [0,1]

```python
import numpy as np

X_20 = np.random.uniform(0,1,20)
```

• Construct the training set T = { (x1,y1),(x2,y2),......,(x20,y20)} using the relation Yi = sin (2 π xi) + εi where εi ~ N(0,0.25)

```python
e_20 = np.random.normal(0,0.25,20)

Y_20 = np.sin(2*3.14*X_20)+e_20

Training_set = list(zip(X_20,Y_20))

Training_set
```

• In the similar way construct a testing set of size 50 i.e. Test = {(x’1,y’1),(x’2,y’2),......,(x’50,y’50)}

```python
X_50 = np.random.uniform(0,1,50)
e_50 = np.random.normal(0,0.25,50)
Y_50 = np.sin(2*3.14*X_50)+e_50

Testing_set = list(zip(X_50,Y_50))

Testing_set
```

Estimate the Least Square polynomial regression model of order M = 1,2, 3, 9, using the training set T.

```python
def A_matrix(order,X):
    m = order+1
    n = len(X)
    A = np.zeros((n,m))
    for i in range(len(X)):
        for j in range(m):
            A[i,j] = X[i]**j
    return A

def LeastSqPolynomialRegression(order,training_set):
    X = [i for i,j in training_set]
    Y = [j for i,j in training_set]
    A = A_matrix(order,X)
    ans = np.dot(np.linalg.inv(np.dot(A.T,A)),np.dot(A.T,Y))
    return ans

#M = 1
ans_m1 = LeastSqPolynomialRegression(1,Training_set)
print('\nLeast Square polynomial regression model of order M = 1: \n', ans_m1)

#M = 2
ans_m2 = LeastSqPolynomialRegression(2,Training_set)
print('\nLeast Square polynomial regression model of order M = 2: \n', ans_m2)

#M = 3
ans_m3 =  LeastSqPolynomialRegression(3,Training_set)
print('\nLeast Square polynomial regression model of order M = 3: \n', ans_m3)

#M = 9
ans_m9 = LeastSqPolynomialRegression(9,Training_set)
print('\nLeast Square polynomial regression model of order M = 9: \n', ans_m9)
```

• List the value of coefficients of estimated polynomial regression models for each case.

```python
def list_values(ans):
    for i in range(len(ans)):
        print(f'b{i} : ',ans[i])


#M = 1
print('\nValue of coefficients of estimated polynomial regression models for M = 1: ')
list_values(ans_m1)

#M = 2
print('\nValue of coefficients of estimated polynomial regression models for M = 2: ')
list_values(ans_m2)

#M = 3
print('\nValue of coefficients of estimated polynomial regression models for M = 3: ')
list_values(ans_m3)

#M = 9
print('\nValue of coefficients of estimated polynomial regression models for M = 9: ')
list_values(ans_m9)
```

• Obtain the prediction on testing set and compute the RMSE for polynomial regression models for order M =1,2,3 and 9.

```python
def predicted(order, testing_set, ans):
    X = [i for i,j in testing_set]
    m = order+1
    predicted = []
    for k in range(len(X)):
        sum = 0
        for i in range(m):
            temp = ans[i]*(X[k]**i) 
            sum = sum + temp
        predicted.append(sum)
    return predicted

def RMSE(predicted,testing_set):
    actual = [j for i,j in testing_set]
    n = len(actual)
    RMSE = 0
    for i in range(n):
        temp = (predicted[i]-actual[i])**2
        RMSE = RMSE + temp
    RMSE = (RMSE/n)**0.5
    return RMSE

```

```python

#M = 1
pred_training_m1 = predicted(1,Training_set,ans_m1)
rmse_training_m1 = RMSE(pred_training_m1,Training_set)
print('\nTrain RMSE for polynomial regression models for order M = 1 : ', rmse_training_m1)

#M = 2
pred_training_m2 = predicted(2,Training_set,ans_m2)
rmse_training_m2 = RMSE(pred_training_m2,Training_set)
print('\nTrain RMSE for polynomial regression models for order M = 2 : ', rmse_training_m2)

#M = 3
pred_training_m3 = predicted(3,Training_set,ans_m3)
rmse_training_m3 = RMSE(pred_training_m3,Training_set)
print('\nTrain RMSE for polynomial regression models for order M = 3 : ', rmse_training_m3)

#M = 9
pred_training_m9 = predicted(9,Training_set,ans_m9)
rmse_training_m9 = RMSE(pred_training_m9,Training_set)
print('\nTrain RMSE for polynomial regression models for order M = 9 : ', rmse_training_m9)
```

```python
#M = 1
pred_m1 = predicted(1,Testing_set,ans_m1)
rmse_m1 = RMSE(pred_m1,Testing_set)
print('\nTest RMSE for polynomial regression models for order M = 1 : ', rmse_m1)

#M = 2
pred_m2 = predicted(2,Testing_set,ans_m2)
rmse_m2 = RMSE(pred_m2,Testing_set)
print('\nTest RMSE for polynomial regression models for order M = 2 : ', rmse_m2)

#M = 3
pred_m3 = predicted(3,Testing_set,ans_m3)
rmse_m3 = RMSE(pred_m3,Testing_set)
print('\nTest RMSE for polynomial regression models for order M = 3 : ', rmse_m3)

#M = 9
pred_m9 = predicted(9,Testing_set,ans_m9)
rmse_m9 = RMSE(pred_m9,Testing_set)
print('\nTest RMSE for polynomial regression models for order M = 9 : ', rmse_m9)
```
